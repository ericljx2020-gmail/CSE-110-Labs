## Background and Prior Work

Since the onset of the COVID-19 pandemic, remote work has dramatically reshaped workplace dynamics, impacting both productivity and mental well-being. The shift from on-site to remote or hybrid work setups has introduced new stressors and benefits. Existing research suggests a complex relationship between work location and mental health outcomes. For instance, Shimura et al. (2021) found that remote work can alleviate physical and psychological stress but noted that it may lead to presenteeism, where employees feel pressured to work despite health concerns. Similarly, Perry et al. (2023) identified that remote workers face more family-related interruptions, potentially raising stress levels.

Despite these findings, previous studies primarily focus on the general population without distinguishing by industry or demographic. This limits their applicability and ability to produce effective target solutions for mental health issues to specific fields of work and industries. Our study expands on this by examining specific demographics (age, gender) within particular industries (tech and finance), offering more tailored insights for companies aiming to improve their support systems. By focusing on demographic and industry-specific factors, our research could reveal underlying trends and contribute to targeted mental health policies.

# Data cleaning & preprocessing

```

import pandas as pd

mental_health_data = pd.read_csv("Impact_of_Remote_Work_on_Mental_Health.csv")

#1.Drop unnecessary columns
columns_to_keep = [
    "Age", "Gender", "Industry", "Work_Location", "Stress_Level",
    "Social_Isolation_Rating", "Work_Life_Balance_Rating", "Mental_Health_Condition",
    "Satisfaction_with_Remote_Work"
]
mental_health_data = mental_health_data[columns_to_keep]

# 2. Rename industries
# Change all industries not labeled "IT" to "Business"
mental_health_data["Industry"] = mental_health_data["Industry"].apply(lambda x: "Business" if x != "IT" else x)

# 3. Filter relevant age groups (Young: 25-35, Mid-Age: 35-45)
mental_health_data = mental_health_data[(mental_health_data["Age"] >= 25) & (mental_health_data["Age"] <= 45)]

# 4. Handle NaN values
# - Fill missing "Stress_Level" and "Social_Isolation" with "Unknown"
mental_health_data["Stress_Level"].fillna("Unknown", inplace=True)
mental_health_data["Social_Isolation_Rating"].fillna("Unknown", inplace=True)

# - Fill missing "Work_Life_Balance" and "Mental_Health_Condition" with "Unknown"
mental_health_data["Work_Life_Balance_Rating"].fillna("Unknown", inplace=True)
mental_health_data["Mental_Health_Condition"].fillna("Unknown", inplace=True)

# Map Satisfaction_with_Remote_Work to numeric
satisfaction_mapping = {"Unsatisfied": 1, "Neutral": 2, "Satisfied": 3}
mental_health_data["Satisfaction_with_Remote_Work"] = mental_health_data["Satisfaction_with_Remote_Work"].map(satisfaction_mapping)

# Ensure Social_Isolation_Rating and Work_Life_Balance_Rating are numeric
mental_health_data["Social_Isolation_Rating"] = pd.to_numeric(mental_health_data["Social_Isolation_Rating"], errors="coerce")
mental_health_data["Work_Life_Balance_Rating"] = pd.to_numeric(mental_health_data["Work_Life_Balance_Rating"], errors="coerce")

# Final preprocessed DataFrame
print("First 10 rows of Remote Work and Mental Health Dataset:")
display(mental_health_data.head(10))
```
